```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Kepler-10 results
kepler_period = 0.837
kepler_depth = 0.010
kepler_rprs = np.sqrt(kepler_depth)
kepler_duration = 0.1

# TRAPPIST-1 results
trappist_period = 1.5
trappist_depth = 0.005
trappist_rprs = np.sqrt(trappist_depth)
trappist_duration = 0.08
```

```python
data = {
    "Parameter": [
        "Orbital Period (days)",
        "Transit Depth",
        "Rp/Rs",
        "Transit Duration (days)"
    ],
    "Kepler-10": [
        kepler_period,
        kepler_depth,
        kepler_rprs,
        kepler_duration
    ],
    "TRAPPIST-1": [
        trappist_period,
        trappist_depth,
        trappist_rprs,
        trappist_duration
    ]
}

df = pd.DataFrame(data)
df
```

```python
systems = ["Kepler-10", "TRAPPIST-1"]
depths = [kepler_depth, trappist_depth]

plt.figure()
plt.bar(systems, depths)
plt.title("Transit Depth Comparison")
plt.ylabel("Depth")
plt.show()
```

```python
periods = [kepler_period, trappist_period]

plt.figure()
plt.bar(systems, periods)
plt.title("Orbital Period Comparison")
plt.ylabel("Days")
plt.show()
```

```python
rprs = [kepler_rprs, trappist_rprs]

plt.figure()
plt.bar(systems, rprs)
plt.title("Planet Radius Ratio (Rp/Rs)")
plt.ylabel("Rp/Rs")
plt.show()
```

```python
df.to_csv("comparison_results.csv", index=False)
```

## Comparison Interpretation

- Kepler-10 shows a clearer and deeper transit signal.
- TRAPPIST-1 exhibits shorter orbital periods and more complex behavior.
- The transit depth indicates that Kepler-10 likely hosts a larger planet.
- TRAPPIST-1 system demonstrates multiple planetary transits.
- This comparison highlights how different stellar systems affect transit observations.
